## **Introduction**

The concepts of reciprocal space and the reciprocal lattice were first introduced by [Paul Peter Ewald](https://en.wikipedia.org/wiki/Paul_Peter_Ewald) (German crystallographer and physicist) in 1921 [[Annalen der Physik, 369, 3, 253 (1921)]](https://onlinelibrary.wiley.com/doi/10.1002/andp.19213690304). The concept of reciprocal space is extremely helpful in understanding X-ray diffraction. It is therefore one of the core concepts in solid-state physics.

In order to construct the lattice of a given crystal structure in reciprocal space, it is first necessary to introduce a method of constructing the real space lattice. The Wigner–Seitz construction is one such method of constructing the real space primitive cell of a crystal. The counterpart (or "dual") of this Wigner-Seitz primitive cell in reciprocal space is called the 1st Brillouin zone.

## **Reciprocal space and reciprocal lattice**

Let's define $\vec{a}_1$, $\vec{a}_2$ and $\vec{a}_3$ as a set of primitive vectors for the direction of the lattice. The lattice vector $\vec{R}_l$ can be written as:

$$\vec{R}_l = (l_1\vec{a}_1 + l_2\vec{a}_2 + l_3\vec{a}_3)$$

The volume of the lattice cell is computed as:

$$\Omega = \vec{a}_1 \cdot (\vec{a}_2 \times \vec{a}_3)$$

Now, let's define another set of primitive vectors: $\vec{b}_1$, $\vec{b}_2$, $\vec{b}_3$. If they satisfy the following relationship with $\vec{a}_1$, $\vec{a}_2$, $\vec{a}_3$:

$$\vec{a}_i \cdot \vec{b}_j = 2\pi \delta_{ij} $$

, where $\delta_{ij}$ is the Dirac delta function, which equals to 1 when $i=j$ and equals
to 0 when $i \neq j$, then both sets of primitive vectors are reciprocal to each other. The primitive vectors $\vec{b}_1$, $\vec{b}_2$, $\vec{b}_3$ can be constructed as:

$$\vec{b}_1 = 2\pi \frac{\vec{a}_2 \times \vec{a}_3}{\vec{a}_1\cdot (\vec{a}_2 \times \vec{a}_3)}$$
$$\vec{b}_2 = 2\pi \frac{\vec{a}_3 \times \vec{a}_1}{\vec{a}_1\cdot (\vec{a}_2 \times \vec{a}_3)}$$
$$\vec{b}_3 = 2\pi \frac{\vec{a}_1 \times \vec{a}_2}{\vec{a}_1\cdot (\vec{a}_2 \times \vec{a}_3)}$$

## **Wigner–Seitz cell**

For solid-state systems, there are many ways to construct the primitive cell
according to the symmetry of the lattice.
The most common method is referred to as the Wigner-Seitz construction, which is named after two physicists: [Eugene Wigner](https://en.wikipedia.org/wiki/Eugene_Wigner) and 
[Frederick Seitz](https://en.wikipedia.org/wiki/Frederick_Seitz).
The Wigner-Seitz cell associated with a given point in a crystalline lattice is defined to be the locus of points in real space which are located nearer to that point than any other lattice point in the crystal under consideration.  

Let us consider, as an example, the steps in the construction for the 2D case. For a 2D lattice grid (as shown in Figure 1), connect one random lattice point to all its nearest 
neighbor lattice points (all the blue line). Then bisect each blue line with a new line (the red line). The smallest polyhedron generated by the intersection of these new red lines is the boundary of the Wigner-Seitz cell. An animation of the method
is shown in the right of Figure 1. We can use the same method to construct the Wigner-Seitz cell in three dimensional space.

## **Brillouin zone**

The concept of a Brillouin zone was first developed by French physicist 
[Léon Brillouin](https://en.wikipedia.org/wiki/L%C3%A9on_Brillouin).
The 1st Brillouin zone is the Wigner-Seitz primitive cell transformed to **reciprocal space**.
As illustrated in Figure 2, the pink area shows the 1st Brillouin zone. The second Brillouin zone
is constructed by drawing perpendicular bisectors of the reciprocal lattice vectors to the second 
nearest lattice points. The region corresponding to the second Brillouin zone is the polyhedron resulting polyhedron with the 
area of the 1st Brillouin zone removed, which is the yellow area in the Figure 2. The green and red areas together are the 3rd Brillouin zone.
